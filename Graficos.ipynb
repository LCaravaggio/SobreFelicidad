```python
# libraries
import matplotlib.pyplot as plt
import seaborn as sns

# data
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/LCaravaggio/SobreFelicidad/main/datos/datos2.csv", sep=";")
```

```python
# use the scatterplot function to build the bubble map
p1= sns.scatterplot(data=df, x=df.GDPpc, y=df.WHR, legend=False)
p1.set(xlim=(0, 150000))
p1.set(ylim=(2, 8))

fig = plt.gcf()
fig.set_size_inches(12, 8)

for line in range(0,df.shape[0]):
     if df.annotation[line] ==1: 
       p1.text(df.GDPpc[line] +0.1, df.WHR[line] , df.Name[line] , horizontalalignment='left', size='medium', color='black')

# show the graph
plt.show()
```

```python
p2=sns.scatterplot(data=df, x=df.Gini, y=df.WHR, legend=False)
p2.set(xlim=(20, 70))
p2.set(ylim=(2, 8))

fig = plt.gcf()
fig.set_size_inches(12, 8)

for line in range(0,df.shape[0]):
     if df.annotation[line]==1: 
       p2.text(df.Gini[line]+0.1, df.WHR[line]-0.1, df.Name[line] , horizontalalignment='left', size='medium', color='black')

plt.show()
```
