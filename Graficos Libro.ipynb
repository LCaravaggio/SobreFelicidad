```python
# libraries
import matplotlib.pyplot as plt
import seaborn as sns

# data
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/LCaravaggio/SobreFelicidad/main/datos/datos2.csv", sep=";")
```

```python
from adjustText import adjust_text

# use the scatterplot function to build the bubble map
p1= sns.scatterplot(data=df, x=df.GDPpc, y=df.WHR22, legend=False)
p1.set(xlim=(0, 150000))
p1.set(ylim=(2, 8))

fig = plt.gcf()
fig.set_size_inches(12, 8)

texts = []
for line in range(0,df.shape[0]):
     if df.annotation2[line] ==1: 
       texts.append(p1.text(df.GDPpc[line], df.WHR22[line], df.Name[line]))
       
adjust_text(texts, ha='center', va='center')
# show the graph
plt.show()
```

```python
p2=sns.scatterplot(data=df, x=df.Gini, y=df.WHR22, legend=False)
p2.set(xlim=(20, 70))
p2.set(ylim=(2, 8))

fig = plt.gcf()
fig.set_size_inches(12, 8)

texts = []
for line in range(0,df.shape[0]):
     if df.annotation2[line]==1: 
       texts.append(p2.text(df.Gini[line], df.WHR22[line], df.Name[line]))

adjust_text(texts, ha='center', va='center')
plt.show()
```
